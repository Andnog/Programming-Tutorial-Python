# <font color='#ee6c4d'>**Word guesser**</font>

Write a program that prints a riddle and then asks you to guess what word it refers to. The word must be valid if it is spelled correctly and it doesn't matter if the whole word or sections of the word are in upper and lower case, it must be considered valid.

<br>

The riddle:

` I’m tall when I’m young, and I’m short when I’m old. What am I?`

The word to guess:

`candle`

<br>

**Example of valid words**:

Word to guess: candle

Words valid:
* `"candle"`
* `"  candle"`
* `"Candle"`
* `"CANDLE"`
* `"canDLE"`
* `"CANdle"`
* etc

Words invalid:
* `"C4ndle"`
* `"C4ndl3"`
* `"sail"`
* `"cand"`
* etc

<br>

#### **Input:**

* word: Given by the user 

<br>

#### **Output:**

If the word were guessed print: `Excelent, the word is candle`, otherwise print `Nice try but it is not the word`.

<br>

> Hint: Print first the riddle and then aks the word

````python
# Solution
print("Riddle: I’m tall when I’m young, and I’m short when I’m old. What am I?")
word = input("The answer is? (Only one word): ")

word = word.lower()
word = word.strip()

if word == "candle":
  print("Excelent, the word is candle")
else:
  print("Nice try but it is not the word")
````

# <font color='#ee6c4d'>**CreditLine**</font>

Write a program that asks for three transaction amounts, averages those amounts and returns a credit line amount according to the average. The program must consider the following conditions: the transaction amounts must be rounded to 2 decimal places, the credit line amount must be rounded to hundreds, and it cannot generate a credit line if the average is less than 75.

<br>

#### **Input:**

* transaction_amount_1: Given by the user.
* transaction_amount_2: Given by the user.
* transaction_amount_3: Given by the user.

<br>

#### **Output:**

Print if the amount is valid: `"Congratulation, your credit amount is $<creditline_amount>"`, otherwise print: `"Sorry, the transaction amount is not sufficient"`

<br>

> Hint: Use round

````python
# Solution
transaction_amount_1 = float(input("Give me the first transaction amount: "))
transaction_amount_2 = float(input("Give me the second transaction amount: "))
transaction_amount_3 = float(input("Give me the third transaction amount: "))

transaction_amount_1 = round(transaction_amount_1, 2)
transaction_amount_2 = round(transaction_amount_2, 2)
transaction_amount_3 = round(transaction_amount_3, 2)

mean_transaction = (transaction_amount_1 + transaction_amount_2 + transaction_amount_3) / 3

if mean_transaction <= 75:
  print("Sorry, the transaction amount is not sufficient")
else:
  total_creditline = round(mean_transaction, -2)
  print("Congratulation, your credit amount is $", total_creditline)
````

# <font color='#ee6c4d'>**Body Mass Index (BMI)**</font>

Write a program that calculates the body mass index based on the person's weight and height, and print the interpretation of the result. The interpretation is given by the following table:

<br>

<table> <thead> <tr>
   <th>Range</th>
   <td align="center"><18.5</td>
   <td align="center">18.5 - 24.9</td>
   <td align="center">25 - 29.9</td>
   <td align="center">30 - 34.9</td>
   <td align="center">>35</td>
</tr> </thead>
<tbody> 
 <tr>
  <td align="center"></td>
  <td align="center">Underweight</td>
  <td align="center">Normal</td>
  <td align="center">Overweight</td>
  <td align="center">Obese</td>
  <td align="center">Extremely obese</td>
 </tr>
</tbody></table>

<br>

#### **Formula:**

$BMI = \frac{weight\_in\_kg}{(height\_in\_m)^2}$

<br>

#### **Input:**

* weight: Given by the user. In Kilograms (Kg)
* height: Given by the user. In Kilograms (cm)

<br>

#### **Output:**

Print: `"Your BMI is <result> and is <range_result>"`

<br>

> Hint: Convert cm to m

> Hint: Use round in BMI result. (2 decimals places)

````python
# Solution
weight = float(input("Give me your weight in kg: "))
height_in_cm = float(input("Give me your height in cm: "))


height_in_m = height_in_cm / 100
bmi = weight / (height_in_m**2)
bmi = round(bmi, 2)

if bmi < 18.5:
  print("Your BMI is " + str(bmi) + " and is underweight" )
elif 18.5 <= bmi <= 24.9:
  print("Your BMI is " + str(bmi) + " and is normal" )
elif 25 <= bmi <= 29.9:
  print("Your BMI is " + str(bmi) + " and is overweight" )
elif 30 <= bmi <= 34.9:
  print("Your BMI is " + str(bmi) + " and is obese" )
else:
  print("Your BMI is " + str(bmi) + " and is extremely obese" )
````

# <font color='#ee6c4d'>**Complete triangle star**</font>

Write a program that performs a complete triangle pattern with the `*` symbol of n levels, the levels depend on the given number.

<br>

#### **Input:**

* number: Given by the user

<br>

#### **Output:**

Complete triangular pattern.

<br>

> Hint: Use nested `for` and double `for`

<br>

#### **Example:**

```
 n = 5

 *
 * *
 * * *
 * * * *
 * * * * *
 * * * *
 * * *
 * *
 *
```

````python
# Solution 1
n = int(input("Give me a number: "))

for i in range(n):
  symbol = ""
  for j in range(i+1):
    symbol += " *"
  print(symbol)

for i in range(1,n):
  symbol = ""
  for j in range(n-i):
    symbol += " *"
  print(symbol)
````

````python
# Solution 2
n = 5

for i in range(n):
  print((i+1)*" *")

for i in range(1, n):
  print((n-i)*" *")
````

# <font color='#ee6c4d'>**Diamond star**</font>

Write a program that performs a diamon pattern with the `*` symbol of n levels, the levels depend on the given number.

<br>

#### **Input:**

* number: Given by the user

<br>

#### **Output:**

Diamond pattern.

<br>

> Hint: Use nested `for` and double `for`

<br>

#### **Example:**

```
n = 4

    *
   ***
  *****
 *******
*********
 *******
  *****
   ***
    *
```

````python
# Solution
number = int(input("Give me a number: "))

for i in range(number-1, 0, -1):
  for j in range(i+1):
    print(" ", end=" ")

  index = number - 1
  odd = ((index - i)*2)+1
  for j in range(odd):
    print("*", end=" ")    
  print()

for i in range(number):
  for j in range(i+1):
    print(" ", end=" ")

  odd = ((index - i)*2)+1
  for j in range(odd):
    print("*", end=" ")
  print()
````
