# <font color='#ee6c4d'>**List operations**</font>

You have been given a list of numbers and you need to perform some operations on it.

1. Create a list of integers with at least 10 elements.
1. Print the entire list.
1. Print the first element of the list.
1. Print the last element of the list.
1. Print the elements from index 2 to index 5 (inclusive).
1. Print the elements from index 5 to the end of the list.
1. Append a new element to the end of the list.
1. Insert a new element at index 3.
1. Remove the element at index 6.
1. Reverse the order of the elements in the list.
1. Sort the elements in the list in ascending order.
1. Sort the elements in the list in descending order.
1. Print the length of the list.
1. Create a new list that contains only the unique elements from the original list.

<br>

#### **Output:**

```
  Original list: [7, 2, 4, 1, 8, 6, 3, 5, 9, 2]

  Entire list: [7, 2, 4, 1, 8, 6, 3, 5, 9, 2]
  First element: 7
  Elements from index 2 to 5: [4, 1, 8, 6]
  Elements from index 5 to the end: [6, 3, 5, 9, 2]
  List after appending 10: [7, 2, 4, 1, 8, 6, 3, 5, 9, 2, 10]
  List after inserting 11 at index 3: [7, 2, 4, 11, 1, 8, 6, 3, 5, 9, 2, 10]
  List after removing element at index 6: [7, 2, 4, 11, 1, 8, 3, 5, 9, 2, 10]
  List after reversing order: [10, 2, 9, 5, 3, 8, 1, 11, 4, 2, 7]
  List sorted in ascending order: [1, 2, 2, 3, 4, 5, 7, 8, 9, 10, 11]
  List sorted in descending order: [11, 10, 9, 8, 7, 5, 4, 3, 2, 2, 1]
  Length of the list: 11
  List with unique elements: [11, 10, 9, 8, 7, 5, 4, 3, 2, 1]
```

````python
original_list = [7, 2, 4, 1, 8, 6, 3, 5, 9, 2]

print("Entire list:", original_list)
print("First element:", original_list[0])
print("Elements from index 2 to 5:", original_list[2:6])
print("Elements from index 5 to the end:", original_list[5:])
original_list.append(10)
print("List after appending 10:", original_list)
original_list.insert(3, 11)
print("List after inserting 11 at index 3:", original_list)
del original_list[6]
print("List after removing element at index 6:", original_list)
original_list.reverse()
print("List after reversing order:", original_list)
original_list.sort(reverse=False)
print("List sorted in ascending order:", original_list)
original_list.sort(reverse=True)
print("List sorted in descending order:", original_list)
print("Length of the list:", len(original_list))

unique_list = []
for item in original_list:
  if item not in unique_list:
    unique_list.append(item)
print("List with unique elements:", unique_list)
````

````python
org_list = [1,22,44,54,36,29,51,1,22,98,58,54]

sum_total = 0
for i in org_list:
  sum_total += i

print("The sum of all elements:", sum_total)

even_num = []
for i in org_list:
  if i%2 == 0:
    even_num.append(i)

print("List of even elements:", even_num)

largest_num = 0
for i in org_list:
  if largest_num < i:
    largest_num = i

print("Largest element:", largest_num)

smallest_num = 999999
for i in org_list:
  if smallest_num > i:
    smallest_num = i

print("Smallest element:", smallest_num)

unique_list = []
for item in org_list:
  if item not in unique_list:
    unique_list.append(item)
print("List of unique elements:", unique_list)

squared_list = []
for item in org_list:
  squared_list.append(item**2)
print("List with elements squared:", squared_list)
````
