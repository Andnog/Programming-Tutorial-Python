# <font color='#ee6c4d'>**Math program**</font>

Create a program with the following instructions:

1. Write a function called calculate_sum that takes in two parameters, num1 and num2, and returns their sum.
1. Write a function called calculate_product that takes in two parameters, num1 and num2, and returns their product.
1. Write a function called calculate_average that takes in a list of numbers and returns their average.
1. Write a function called find_maximum that takes in a list of numbers and returns the maximum number in the list.
1. Write a function called find_minimum that takes in a list of numbers and returns the minimum number in the list.
1. Write a program that prompts the user to enter two numbers, calculates and prints their sum and product using the calculate_sum and calculate_product functions.
1. Write a program that prompts the user to enter a list of numbers, calculates and prints their average, maximum, and minimum using the calculate_average, find_maximum, and find_minimum functions.
1. Add error handling to your programs to handle situations where the user enters invalid input (such as non-numeric values).

````python
# Solution
# Define functions
def calculate_sum(num1, num2):
    return num1 + num2

def calculate_product(num1, num2):
    return num1 * num2

def calculate_average(numbers):
    return sum(numbers) / len(numbers)

def find_maximum(numbers):
    return max(numbers)

def find_minimum(numbers):
    return min(numbers)

# Program to calculate sum and product
num1 = input("Enter the first number: ")
num2 = input("Enter the second number: ")

try:
    num1 = float(num1)
    num2 = float(num2)
    print("Sum:", calculate_sum(num1, num2))
    print("Product:", calculate_product(num1, num2))
except ValueError:
    print("Invalid input. Please enter numeric values.")

# Program to calculate average, maximum, and minimum
numbers = input("Enter a list of numbers separated by commas: ")

try:
    numbers_list = [float(x) for x in numbers.split(",")]
    print("Average:", calculate_average(numbers_list))
    print("Maximum:", find_maximum(numbers_list))
    print("Minimum:", find_minimum(numbers_list))
except ValueError:
    print("Invalid input. Please enter numeric values separated by commas.")
````

# <font color='#ee6c4d'>**Store inventory**</font>

Create a program with the following instructions:

1. Create a dictionary called `inventory` with items as the keys and their quantities and prices as nested dictionaries.
1. Create a list called `cart` to store the items the customer wants to purchase.
1. Create a function called `calculate_total` that takes in `inventory` and `cart` as arguments and returns the total cost of the items in the cart. The function should loop through the items in the cart, look up their quantity and price in the `inventory` dictionary, and calculate the subtotal for each item. The subtotals should then be added up to get the total cost.
1. Call the `calculate_total` function with `inventory` and `cart` as arguments, and print the result. 

````python
# Solution
inventory = {
    "apples": {"quantity": 5, "price": 0.5},
    "bananas": {"quantity": 3, "price": 0.25},
    "oranges": {"quantity": 2, "price": 0.75}
}

cart = ["apples", "bananas", "oranges", "apples"]

def calculate_total(inventory, cart):
    total_cost = 0
    for item in cart:
        if item in inventory:
            quantity = inventory[item]["quantity"]
            price = inventory[item]["price"]
            subtotal = quantity * price
            total_cost += subtotal
    return total_cost

print("Total cost:", calculate_total(inventory, cart))
````

````python
# Solution

print("Welcome to attendance program asistance!")
students_attendance = []

def menu_validation(valid_options, input_message, option):
  while(option not in valid_options):
    print()
    print("No a valid option. Try again. The valid options are:", end=" ")
    for option in valid_options:
      print(option.upper() + ", ", end="")
    print()
    option = input(input_message).lower()
  return option.lower()

def print_general_menu(title, content):
  header = "-"*10 + " " + title + " " + "-"*10

  print()
  print(header)
  print(content.replace("  ", ""))
  print("-"*len(header))

def print_menu():
  menu =  """
              1 - Show student's information
              2 - Show student's assistance
              3 - Show number of total students
              4 - Modify student's information
              5 - Delete student's information
              6 - Quit program
          """
  print_general_menu("MENU", menu)

def print_student_information(student):
  name, lastname, assistance = student
  print()
  print("Student information")
  print("Name:",name.title())
  print("Lastname:",lastname.title())
  print("Assistance:",assistance.title())

def print_student_name_info(name_info, names_list):
  print()
  count = 1
  for name in names_list:
      print("Student", count, name_info,":", name.title())
      count += 1

def get_student_list_info(students_attendance, attribute):
  dict_attr = {"name": 0, "lastname": 1, "assistance": 2, "fullname": 3}

  info_list = []
  index = dict_attr[attribute]
  for student in students_attendance:
    if index <= 2:
      info_list.append(student[index])
    else:
      info_list.append(student[0].lower() + " " + student[1].lower())

  return info_list

assistance_dict = {"p": "Present", "a": "Absent"}
while(True):
    print()
    name = input("Student name and middle name: ")
    lastname = input("Student lastname: ")
    assistance = input("Student assistance (P - present | A - abstent): ").lower()
    valid_options = ["p", "a"]
    assistance = menu_validation(valid_options, "Student assistance: ", assistance)
    students_attendance.append([name, lastname, assistance_dict[assistance]])

    print()
    option = input("Do you want to register other student? (Y/N) ").lower()

    valid_options = ["y", "n"]
    option = menu_validation(valid_options, "Do you want to register other student? (Y/N) ", option)

    if option == 'n':
        break

while(True):
    print_menu()
    option = input("Select an option: ")

    valid_options = ["1","2","3","4","5","6"]
    while(option not in valid_options):
        print()
        print("No a valid option. Try again.")
        option = input("Select an option: ")

    # Show student's information
    if option == "1":
        while(True):
            submenu =   """
                            1 - Show a student information by name
                            2 - Show a student information by position
                            3 - Show all students names
                            4 - Show all students lastnames                        
                            5 - Return to main menu
                        """
            print_general_menu("SUBMENU", submenu)
            suboption = input("Select an option: ")

            valid_options = ["1","2","3","4","5"]
            suboption = menu_validation(valid_options, "Select an option: ", suboption)

            if suboption == "1":
                print()
                name_to_search = input("Which name do you want to search? ")
                is_found = False

                for student in students_attendance:
                    name = student[0]
                    if name.lower() == name_to_search.lower():
                        print_student_information(student)
                        is_found = True

                if not is_found:
                    print("Student not found")

            elif suboption == "2":
                print()
                index_to_search = int(input("Which position do you want to search? "))                

                if 1 <= index_to_search <= len(students_attendance):
                    print_student_information(students_attendance[index_to_search - 1])
                else:
                    print("The position is not valid. Max position:", len(students_attendance))

            elif suboption == "3":
                print()
                names_list = get_student_list_info(students_attendance, "name")

                is_order = input("Would you like to order the names alphabetically? (Y/N) ").lower()

                valid_options = ["y","n"]
                is_order = menu_validation(valid_options, "Select an option: ", is_order)

                if is_order == 'y':
                    is_ascending = input("Do you want ascending or descending? (A/D) ").lower()

                    valid_options = ["a","d"]
                    is_ascending = menu_validation(valid_options, "Select an option: ", is_ascending)

                    if is_ascending == "a":
                        names_list.sort()
                    else:
                        names_list.sort(reverse=True)

                    print_student_name_info("name", names_list)

                else:
                    print_student_name_info("name", names_list)

            elif suboption == "4":
                print()
                lastnames_list = get_student_list_info(students_attendance, "lastname")

                is_order = input("Would you like to order the lastnames alphabetically? (Y/N) ").lower()

                valid_options = ["y","n"]
                is_order = menu_validation(valid_options, "Select an option: ", is_order)

                if is_order == 'y':
                    is_ascending = input("Do you want ascending or descending? (A/D) ").lower()

                    valid_options = ["a","d"]
                    is_ascending = menu_validation(valid_options, "Select an option: ", is_ascending)

                    if is_ascending == "a":
                        lastnames_list.sort()
                    else:
                        lastnames_list.sort(reverse=True)

                    print_student_name_info("lastname", lastnames_list)

                else:
                    print_student_name_info("lastname", lastnames_list)

            elif suboption == "5":
                break

    # Show student's assistance
    elif option == "2":
        print()
        assistance_list = get_student_list_info(students_attendance, "assistance")
        print(assistance_list)
        total_present = assistance_list.count("Present")
        total_abstence = assistance_list.count("Absent")

        print("Total students present:", total_present, "| Total students absent:", total_abstence)

    # Show number of total students
    elif option == "3":
        print()
        print("Total students:", len(students_attendance))
        pass

    # Modify student's information
    elif option == "4":
        print()
        print("To modify student information it is necessary to provide the student's full name")
        fullname_list = get_student_list_info(students_attendance, "fullname")
        name_to_search = input("Write the student's full name: ").lower()

        idx = 0
        for fullname in fullname_list:
            if name_to_search.lower() == fullname.lower():
                break
            idx += 1

        if idx <= len(students_attendance) - 1:
            name = input("Student name and middle name: ")
            lastname = input("Student lastname: ")
            assistance = input("Student assistance (P - present | A - abstent): ").lower()

            valid_options = ["p", "a"]
            assistance = menu_validation(valid_options, "Student assistance: ", assistance)
            students_attendance[idx] = [name, lastname, assistance_dict[assistance]]
        else:
            print("Student not found")

    # Delete student's information
    elif option == "5":
        submenu =   """
                        1 - Delete last student information
                        2 - Delete student information by position
                    """
        print_general_menu("SUBMENU", submenu)
        suboption = input("Select an option: ")

        valid_options = ["1","2"]
        suboption = menu_validation(valid_options, "Select an option: ", suboption)

        if suboption == "1":
            last_element_deleted = students_attendance.pop()
            print("The student information was deleted:")
            print_student_information(last_element_deleted)

        elif suboption == "2":
            index_to_delete = int(input("Which position do you want to delete? "))

            if 0 < index_to_delete <= len(students_attendance):
                element_deleted = students_attendance[index_to_delete - 1]
                del students_attendance[index_to_delete - 1]
                print("The student information in position", index_to_delete, "was deleted:")
                print_student_information(element_deleted)
            else:
                print("The position is not valid. Max position:", len(students_attendance))

    # Quit program      
    elif option == "6":
        print("Bye! ;)")
        break
````

````python
def add_item(cart):
    item = input("Enter the name of the item to add: ")
    quantity = int(input("Enter the quantity to add: "))
    if item in cart:
        cart[item] += quantity
    else:
        cart[item] = quantity
    print(f"{quantity} {item}(s) added to cart.")
    
def remove_item(cart):
    item = input("Enter the name of the item to remove: ")
    if item in cart:
        del cart[item]
        print(f"{item} removed from cart.")
    else:
        print(f"{item} is not in the cart.")
    
def view_cart(cart):
    print("Items in cart:")
    for item, quantity in cart.items():
        print(f"{item}: {quantity}")
        
def checkout(cart, prices):
    total = 0
    print("Cart summary:")
    for item, quantity in cart.items():
        if item.lower() in prices:
          price = prices.get(item.lower(), 0)
        else:
          price = prices.get("other", 0)

        cost = price * quantity
        total += cost
        print(f"{item.lower()}: {quantity} x {price} = ${cost}")
    print(f"Total cost: ${total}")
    
cart = {}
fruits_prices = {
    "apple": 0.5,
    "banana": 0.25,
    "orange": 0.75,
    "grape": 1.0,
    "kiwi": 1.5,
    "other": 1
}
while True:
    print("\nMenu:")
    print("1. View cart")
    print("2. Add item to cart")
    print("3. Remove item from cart")
    print("4. Checkout")
    print("5. Quit")
    choice = input("Enter your choice (1-5): ")
    if choice == "1":
        view_cart(cart)
    elif choice == "2":
        add_item(cart)
    elif choice == "3":
        remove_item(cart)
    elif choice == "4":
        checkout(cart, fruits_prices)
    elif choice == "5":
        break
    else:
        print("Invalid choice. Try again.")
````
